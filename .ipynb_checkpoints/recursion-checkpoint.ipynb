# 부분집합 생성하기

## 원소가 n개인 집합의 모든 부분집합 생성하기

### 1. itertools의 combinations

```python
from itertools import combinations

arr = [1,2,3]
result = []
for i in range(len(arr)+1):
    #Needs type converting (combinations -> list)
    result += list(combinations(arr,i)) 
    
print(result)
```

<b> 시간복잡도 O(n*nCr) <br> -> r에 따라서 탐색해야할 길이가 달라짐

### 2. 단순반복문

```python
arr = [1,2,3]
result = [[]]

for elements in arr:
    length = len(result)
    for i in range(length):
        result.append(result[i]+[elements])
        
print(result)
```

<b> 시간복잡도 O(N*2^N)

### 3. 재귀함수

```python
import copy

result = []
arr = [1,2,3]

def recur(subset, i , arr):
    if i == len(arr): #Recursion escape condition
        result.append(copy.copy(subset))
        #print(f'등록: i={i},subset={subset}')
        return
    else:
        subset.append(arr[i])
        i+=1
        #print(f'i={i},subset={subset}')
        recur(subset,i,arr)
        subset.pop()
        recur(subset,i,arr)
        

recur([],0,arr)
print(result)
```

<B> 시간복잡도 O(N*2^N) <br> -1 원소가 있고 없고, -2 원소가 있고 없고, -3 원소가 있고없고
